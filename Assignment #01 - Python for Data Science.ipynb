# 1. Read the dataset to python environment.

# 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Classification',y='PW', data=df)
plt.title('Variation of PW for each Classification', fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.countplot(x='Classification', data=df)
plt.title("Count of each Classification", fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='SW',y='PL', hue='Classification', data=df)
plt.title("SW-PL Variation for each classification", fontsize=20)
plt.show()
```
